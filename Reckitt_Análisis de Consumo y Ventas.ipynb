```python
# Consulta mediante INNER JOIN
cur.execute('''SELECT 
                    c.DATE,
                    p.BRAND, 
                    p.CATEGORY,
                    s.ATTR1,
                    sl.TOTAL_UNIT_SALES,
                    sl.TOTAL_VALUE_SALES,
                    sl.REGION
                FROM df_product AS p  
                INNER JOIN (SELECT FORMAT, ATTR1 FROM df_segment GROUP BY FORMAT) AS s ON p.FORMAT = s.FORMAT
                INNER JOIN df_sales AS sl ON p.ITEM = sl.ITEM_CODE
                INNER JOIN (SELECT WEEK, DATE FROM df_calendar GROUP BY WEEK) AS c ON sl.WEEK = c.WEEK
            ''')

Reckitt = cur.fetchall()
print(f"Successfully fetched {len(Reckitt)} rows")
```

```python
# Creación de la tabla conbinada de los joins
cur.execute('''
        CREATE TABLE Reckitt (
            DATE DATE,
            BRAND TEXT,
            CATEGORY TEXT,
            ATTR1 TEXT,  
            TOTAL_UNIT_SALES REAL, 
            TOTAL_VALUE_SALES REAL,
            REGION TEXT);''')
```

```python
# Inserción de los datos en la tabla Reckitt
for renglon in range(1, len(Reckitt)):
    cur.execute('''
        INSERT INTO Reckitt (DATE, BRAND, CATEGORY, ATTR1, TOTAL_UNIT_SALES, TOTAL_VALUE_SALES, REGION)
        VALUES (?,?,?,?,?,?,?)
        ''', Reckitt[renglon])
conn.commit()
```

```python
# Revisión de la tabla Reckitt revisado en DB Browser y aquí
cur.execute("SELECT * FROM Reckitt")
Reckitt_result = cur.fetchall()
Reckitt_result
```

```python
# Comenzar índice de primera fila con 1
df = df.reset_index(drop=True); df.index = range(1, len(df) + 1)
print(df.shape)
df
```

# Segmentación de Productos con Clustering
* Aplicación de Clustering con K-Means: Se aplicó el algoritmo K-Means para segmentar los productos en función de características clave como ventas totales.
* Número Óptimo de Clusters: Se utilizó el método del codo para determinar el número óptimo de clusters, logrando identificar patrones de rendimiento.
* Interpretación de los Clusters: Se analizó cada cluster para identificar áreas de alto rendimiento y segmentos con oportunidades de mejora.
* Visualizaciones de Clusters: Se generaron gráficos de dispersión para visualizar los clusters y entender mejor las relaciones entre las diferentes variables.

```python
from sklearn.cluster import KMeans
df = pd.read_csv('Reckitt_df.csv')
df.head()
```

```python
df.info()
```

```python
# Convertir variables categóricas en numéricas
from sklearn.preprocessing import LabelEncoder

label_encoders = {}
for col in df:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = dict(zip(le.classes_, le.transform(le.classes_)))

df.dtypes
```

```python
# Referencia de str e int en BRAND
label_encoders['BRAND']
```

```python
# Referencia de str e int en REGION
label_encoders['REGION']
```

```python
# Referencia de str e int en CATEGORY
label_encoders['CATEGORY']
```

```python
# Selección de dos variables para analizar, en este caso BRAND Y TOTAL_VALUE_SALES
fields = df.iloc[:,[1,5]].values
print(fields)
```

```python
# Lista vacía para que guarde dinámico los resultados
wcss = []

# Hacer pruebas de hasta 11 clusters para observar que es lo que nos conviene manejar
for index in range(1,11):
    kmeans = KMeans(n_clusters = index, init = 'k-means++', random_state = 2)
    kmeans.fit(fields)

    wcss.append(kmeans.inertia_)

# Graficar "Codo de Jambu"
sns.set()
plt.plot(range(1,11), wcss)
plt.title('K-Means Clustering')
#plt.xlabel('Número de clusters')
#plt.ylabel('WCSS')

# Indicar el número óptimo 
k_optimo = 2
plt.axvline(x=k_optimo, color='red', linestyle='--', label=f'Codo en k = {k_optimo}')
plt.legend()
plt.show()
```

**La clasificación de los grupos por clúster indica que se separaron por los productos/marcas más vendidos y los menos vendidos, mostrando que se dividen en 2 clústers principalmente.**

```python
# Crear gráfica de series
plt.figure(figsize=(12, 4))
plt.plot(series)
plt.title("Serie temporal: TOTAL_VALUE_SALES")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# remover fechas duplicadas en el índice
if not df.index.is_unique:
    print(f"Se detectaron {df.index.duplicated().sum()} fechas duplicadas. Se agregan promedios por fecha.")
    # Agrupamos por fecha y tomamos el promedio de TOTAL_VALUE_SALES
    df = df.groupby(df.index).mean()
    series = df["TOTAL_VALUE_SALES"]
```

```python
# División entrenamiento/prueba (80% train, 20% test)
n = len(series)
train_size = int(n * 0.8)
train, test = series.iloc[:train_size], series.iloc[train_size:]
print(f"Tamaño total: {n}, entrenamiento: {len(train)}, prueba: {len(test)}")
```

```python
# Búsqueda de mejor modelo ARIMA (p,d,q) por AIC
import warnings
warnings.filterwarnings('ignore')
# Como la serie ya la consideraste estacionaria -> d = 0
d = 0
candidate_orders = [(1,0,0), (0,0,1), (1,0,1), (2,0,0), (0,0,2), (2,0,1), (1,0,2)]
p_max = 5
q_max = 5
best_aic = np.inf
best_order = None
best_model = None
print("Probando combinaciones ARIMA comunes...")
for order in candidate_orders:
    try:
        model = sm.tsa.SARIMAX(train, order=order,
                               enforce_stationarity=True,
                               enforce_invertibility=True)
        res = model.fit(disp=False)
        print(f"ARIMA{order} -> AIC = {res.aic:.2f}")
        if res.aic < best_aic:
            best_aic = res.aic
            best_order = order
            best_model = res
    except Exception as e:
        print(f"Falló ARIMA{order}: {e}")
        continue

if best_order is None:
    raise RuntimeError("No se encontró un modelo válido entre las combinaciones comunes.")

print(f"\n Mejor modelo encontrado: ARIMA{best_order} con AIC = {best_aic:.2f}")
print(best_model.summary())
```

```python
# Predicciones
# Índices numéricos seguros
start_train = 0
end_train = len(train) - 1
start_forecast = end_train + 1
end_forecast = end_train + len(test)

# Fitted (in-sample)
fitted_values = best_model.predict(start=start_train, end=end_train)

# Forecast (out-of-sample)
forecast_res = best_model.get_forecast(steps=len(test))
forecast_mean = forecast_res.predicted_mean
forecast_ci = forecast_res.conf_int()

# Combinar resultados en un DataFrame
results_df = pd.DataFrame({"actual": pd.concat([train, test]),
                           "fitted": pd.concat([pd.Series(fitted_values, index=train.index),pd.Series([np.nan]*len(test), index=test.index)]),
                           "forecast": pd.concat([pd.Series([np.nan]*len(train), index=train.index), pd.Series(forecast_mean, index=test.index)])})
```

```python
# Métricas en la prueba
y_true = test.values
y_pred = forecast_mean.values
mse = mean_squared_error(y_true, y_pred)
rmse = math.sqrt(mse)
mape = np.mean(np.abs((y_true - y_pred) / np.where(y_true==0, 1e-8, y_true))) * 100

print("\nMétricas (conjunto de prueba):")
print(f"MSE:  {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAPE: {mape:.2f}%")
```

* **MSE:  294.7217 significa promedio de los errores al cuadrado entre los valores reales y los predichos.**
* **RMSE: 17.1675 significa, en promedio, que el modelo falla en unas 17.17 unidades de TOTAL_VALUE_SALES.**
* **MAPE: 11.19% significa que el modelo predice con un error promedio de 11% respecto al valor real, muy decente para una serie económica o de ventas, sobre todo si no hay estacionalidad fuerte.**
* **Conclusión: con el modelo ARIMA se tiene un buen desempeño: predice razonablemente bien, con errores moderados y consistentes; se puede mejorar añadiendo estacionalidad (SARIMA).**

```python
# Añadir estacionalidad (SARIMA)
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
s = 7
candidate_orders = [((1,0,1), (1,1,1,s)),((1,0,0), (1,1,1,s)),((0,0,1), (1,1,1,s)),((2,0,1), (1,1,0,s)),((1,0,1), (0,1,1,s))]
best_aic = np.inf
best_order = None
best_seasonal_order = None
best_model = None
print("\nProbando combinaciones SARIMA comunes...")
for order, seasonal_order in candidate_orders:
    try:
        model = sm.tsa.statespace.SARIMAX(
            train,
            order=order,
            seasonal_order=seasonal_order,
            enforce_stationarity=False,
            enforce_invertibility=False
        )
        res = model.fit(disp=False)
        print(f"SARIMA{order}x{seasonal_order} -> AIC={res.aic:.2f}")
        if res.aic < best_aic:
            best_aic = res.aic
            best_order = order
            best_seasonal_order = seasonal_order
            best_model = res
    except Exception as e:
        print(f"Falló SARIMA{order}x{seasonal_order}: {e}")

print(f"\n Mejor modelo: SARIMA{best_order}x{best_seasonal_order} con AIC={best_aic:.2f}")
```
